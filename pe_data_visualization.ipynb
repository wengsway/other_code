```python
#import necessary libraries 
import pandas as pd 
import plotly.graph_objects as go

#read excel file 
data1 = pd.read_excel('data.xlsx', sheet_name='Sheet1')

#create a list of x and y coordinates 
x = data1['日期']
y1 = data1['全市场pe'] 
y2 = data1['历史平均']
y3 = data1['五年平均']
y4 = data1['十年平均']
y5 = data1['十年国债市盈率']
y6 = data1['深综指']

#plot the graph
trace0 = go.Scatter(x=x, y=y1, mode='lines', name='全市场pe')
trace1 = go.Scatter(x=x, y=y2, mode='lines', name='历史平均')
trace2 = go.Scatter(x=x, y=y3, mode='lines', name='五年平均')
trace3 = go.Scatter(x=x, y=y4, mode='lines', name='十年平均')
trace4 = go.Scatter(x=x, y=y5, mode='lines', name='十年国债市盈率')
trace5 = go.Scatter(x=x, y=y6, mode='lines', name='深综指', yaxis='y2')
data2 = [trace0, trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(xaxis=dict(rangeselector=dict(buttons=list([dict(count=1,label='1m',step='month',stepmode='backward'),
                   dict(count=6,label='6m',step='month',stepmode='backward'),
                   dict(count=1,label='YTD',step='year',stepmode='todate'),
                   dict(count=1,label='1y',step='year',stepmode='backward'),
                   dict(count=2,label='2y',step='year',stepmode='backward'),
                   dict(count=3,label='3y',step='year',stepmode='backward'),
                   dict(count=5,label='5y',step='year',stepmode='backward'),
                   dict(count=10,label='10y',step='year',stepmode='backward'),
                   dict(count=15,label='15y',step='year',stepmode='backward'),
                   dict(count=20,label='20y',step='year',stepmode='backward'),
                   dict(step='all')])),
                   rangeslider=dict(),
                   type='date'),
                   yaxis=dict(title='等权PE'),
                   yaxis2=dict(title='深综指', overlaying='y',side='right'),
                   legend=dict(orientation="h",xanchor="center", x=0.5, y=0.9999))

fig = go.Figure(data=data2,layout=layout)
fig.update_layout(title={
    'text': 'A股全市场PE与深综指走势对比图',
    'x': 0.5,
    'xanchor': 'center'
    })
fig.add_annotation(text="数据来源：理杏仁",xref="paper", yref="paper",x=1, y=1,showarrow=False)
fig.update_xaxes(tick0=x[0],dtick='M6',showgrid=True,tickformat="%m\n%Y",rangeslider_visible=True)
fig.update_layout(hovermode="x unified")
fig.show()

#generate the web version 
fig.write_html('A股全市场PE与深综指走势对比图.html')
```
